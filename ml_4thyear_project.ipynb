```python
import pandas as pd

# استخراج الأعمدة العددية (باستثناء "Class")
numeric_columns = df.drop(columns=['Class']).select_dtypes(include=['float64', 'int64']).columns

# نسخ البيانات للعمل عليها
df_cleaned = df.copy()

# التعامل مع القيم المتطرفة باستخدام IQR
for col in numeric_columns:
    Q1 = df_cleaned[col].quantile(0.25)  # الربع الأول
    Q3 = df_cleaned[col].quantile(0.75)  # الربع الثالث
    IQR = Q3 - Q1                        # مجال القيم بين الربعين

    # تحديد القيم الحدودية المقبولة
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # استبدال القيم المتطرفة بالحدود
    df_cleaned[col] = df_cleaned[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))

# طباعة معلومات عن البيانات بعد التنظيف
print("Data cleaned. Shape:", df_cleaned.shape)
```

```python
df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
```

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(probability=True),
    'Decision Tree': DecisionTreeClassifier()
}
param_grid = {
'Logistic Regression': {'C': [0.1, 1, 10]},
    'Naive Bayes': {}, # Add an empty dictionary for Naive Bayes since it doesn't usually require hyperparameter tuning
    'SVM': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']},
    'Decision Tree': {'max_depth': [None, 5, 10]},
}

from sklearn.model_selection import train_test_split

# Step 1: Create smaller training data
X_train_small, _, y_train_small, _ = train_test_split(X_resampled, y_resampled, test_size=0.8, random_state=42)

model_results = {}

# Step 2: Loop through the models
for model_name in models:
    print(f"Training {model_name}...")

    # Use smaller dataset only for specific models
    if model_name in ['SVM', 'Decision Tree']:
        train_data, train_labels = X_train_small, y_train_small
    else:
        train_data, train_labels = X_resampled, y_resampled

    # GridSearchCV for hyperparameter tuning
    grid_search = GridSearchCV(estimator=models[model_name], param_grid=param_grid[model_name], cv=3, n_jobs=-1, verbose=1)
    grid_search.fit(train_data, train_labels)

    # Best model after grid search
    best_model = grid_search.best_estimator_

    # Refit the best model on the full dataset for SVM and Decision Tree
    if model_name in ['SVM', 'Decision Tree']:
        best_model.fit(X_resampled, y_resampled)

    # Predicting on the test set
    y_pred = best_model.predict(X_test)
    y_pred_proba = best_model.predict_proba(X_test)[:, 1] if hasattr(best_model, "predict_proba") else None

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy for {model_name}: {acc:.4f}")

    # Store results
    model_results[model_name] = {
        'best_params': grid_search.best_params_,
        'classification_report': classification_report(y_test, y_pred),
        'confusion_matrix': confusion_matrix(y_test, y_pred),
    }

    if y_pred_proba is not None:
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
        model_results[model_name]['roc_curve'] = (fpr, tpr)
        model_results[model_name]['roc_auc'] = auc(fpr, tpr)

    # Plotting the confusion matrix
    cm = model_results[model_name]['confusion_matrix']
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Plotting the ROC Curve
    if y_pred_proba is not None:
        fpr, tpr = model_results[model_name]['roc_curve']
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {model_results[model_name]["roc_auc"]:.2f})', color='blue')
        plt.plot([0, 1], [0, 1], 'k--', color='red')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {model_name}')
        plt.legend(loc='lower right')
        plt.show()

# Step 3: Display results
for model_name, results in model_results.items():
    print(f"Model: {model_name}")
    print(f"Best Parameters: {results['best_params']}")
    print(f"Accuracy: {accuracy_score(y_test, best_model.predict(X_test)):.4f}")
    print(f"Classification Report:\n{results['classification_report']}")
    if 'roc_auc' in results:
        print(f"ROC AUC: {results['roc_auc']}")
    print('-' * 80)
```

```python

from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Best parameters from the base models
best_params = {
    'Logistic Regression': {'C': 10},
    'Naive Bayes': {},  # No hyperparameters
    'SVM': {'C': 10, 'kernel': 'rbf', 'probability': True},
    'Decision Tree': {'max_depth': None}
}

# Initialize base models with best parameters
base_models = {
    'Logistic Regression': LogisticRegression(**best_params['Logistic Regression']),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(**best_params['SVM']),
    'Decision Tree': DecisionTreeClassifier(**best_params['Decision Tree']),
}

# Initialize ensemble models using base model parameters
ensemble_models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=best_params['Decision Tree']['max_depth']),
    'Bagging': BaggingClassifier(estimator=DecisionTreeClassifier(**best_params['Decision Tree']), n_estimators=10), # Change 'base_estimator' to 'estimator'
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, max_depth=best_params['Decision Tree']['max_depth']),
    'XGBoost': XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=best_params['Decision Tree']['max_depth'])
}

# Results storage
ensemble_results = {}

# Evaluate ensemble models
for model_name, model in ensemble_models.items():
    print(f"Training {model_name}...")

    # Train on the resampled dataset
    model.fit(X_resampled, y_resampled)

    # Predict
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy for {model_name}: {acc:.4f}")

    # Store results
    ensemble_results[model_name] = {
        'accuracy': acc,
        'classification_report': classification_report(y_test, y_pred),
        'confusion_matrix': confusion_matrix(y_test, y_pred),
    }

    if y_pred_proba is not None:
        fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
        ensemble_results[model_name]['roc_curve'] = (fpr, tpr)
        ensemble_results[model_name]['roc_auc'] = auc(fpr, tpr)

    # Plot confusion matrix
    cm = ensemble_results[model_name]['confusion_matrix']
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()

    # Plot ROC Curve
    if y_pred_proba is not None:
        fpr, tpr = ensemble_results[model_name]['roc_curve']
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {ensemble_results[model_name]["roc_auc"]:.2f})', color='blue')
        plt.plot([0, 1], [0, 1], 'k--', color='red')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {model_name}')
        plt.legend(loc='lower right')
        plt.show()

# Display results
for model_name, results in ensemble_results.items():
    print(f"Model: {model_name}")
    print(f"Accuracy: {results['accuracy']:.4f}")
    print(f"Classification Report:\n{results['classification_report']}")
    if 'roc_auc' in results:
        print(f"ROC AUC: {results['roc_auc']:.4f}")
    print('-' * 80)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Define base models and meta-learner
base_models = [
    ('rf', RandomForestClassifier(n_estimators=200, random_state=42)),
    ('xgb', XGBClassifier(n_estimators=100, max_depth=best_params['Decision Tree']['max_depth'], random_state=42)),
    # Change 'base_estimator' to 'estimator' in BaggingClassifier
    ('bagging', BaggingClassifier(estimator=DecisionTreeClassifier(max_depth=best_params['Decision Tree']['max_depth']), n_estimators=10, random_state=42)),
]
meta_learner = LogisticRegression(C=best_params['Logistic Regression']['C'], random_state=42)

# Create the stacking classifier
stacking_clf = StackingClassifier(estimators=base_models, final_estimator=meta_learner, cv=5, n_jobs=-1)

# Train the stacking classifier
stacking_clf.fit(X_resampled, y_resampled)

# Evaluate the stacking model on the test set
y_pred = stacking_clf.predict(X_test)
y_pred_proba = stacking_clf.predict_proba(X_test)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, digits=4)
conf_matrix = confusion_matrix(y_test, y_pred)

print("--- Stacking Model Evaluation ---")
print(f"Accuracy: {accuracy:.4f}")
print(f"Classification Report:\n{report}")

# ROC Curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Confusion Matrix Visualization
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Fraud', 'Fraud'], yticklabels=['Not Fraud', 'Fraud'])
plt.title('Confusion Matrix - Stacking')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# ROC Curve Visualization
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})', color='blue')
plt.plot([0, 1], [0, 1], 'k--', color='red')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Stacking Model')
plt.legend(loc='lower right')
plt.show()

# Initialize best_models dictionary if it doesn't exist
best_models = best_models if 'best_models' in locals() else {}

# Add Stacking Model to Best Models
best_models['Stacking'] = stacking_clf

# Save Results
stacking_results = {
    'accuracy': accuracy,
    'classification_report': report,
    'confusion_matrix': conf_matrix,
    'roc_auc': roc_auc,
}

# Display Results
print(f"--- Stacking ---")
print(f"Accuracy: {stacking_results['accuracy']:.4f}")
print(f"ROC AUC: {stacking_results['roc_auc']:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data preparation
models = ["LR", "DT", "SVM", "KNN", "RF"]
accuracy = [99.79, 99.79, 99.79, 99.20, 99.96]  # Example data
roc_auc = [96.36, 89.68, 88.53, 93.00, 97.24]   # Example data

# Convert percentages to a scale of 0-100 for visualization
accuracy_percentage = [acc for acc in accuracy]
roc_auc_percentage = [roc for roc in roc_auc]

# Bar settings
x = np.arange(len(models))  # X-axis positions
width = 0.35  # Width of each bar

# Plot bars
plt.figure(figsize=(8, 6))
plt.bar(x - width/2, accuracy_percentage, width, label="Accuracy (%)", color="mediumpurple")
plt.bar(x + width/2, roc_auc_percentage, width, label="ROC (%)", color="blueviolet")

# Adding labels and title
plt.ylabel("Performance (%)", fontsize=12)
plt.title("Performance Evaluation - Breast Cancer Prediction", fontsize=14)
plt.xticks(x, models, fontsize=10)
plt.ylim(70, 100)  # Set y-axis range for better visualization

# Adding legend
plt.legend(loc="upper right", fontsize=10)

# Annotate values on bars
for i in range(len(models)):
    plt.text(x[i] - width/2, accuracy_percentage[i] + 1, f"{accuracy_percentage[i]:.2f}%", ha="center", fontsize=10)
    plt.text(x[i] + width/2, roc_auc_percentage[i] + 1, f"{roc_auc_percentage[i]:.2f}%", ha="center", fontsize=10)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
